```python
df=spark.read.format("csv").option("header","true").load("s3://proj-nova/Retail_sales/metadata.csv")
```

```python
df.show()
```

```python
def normalize_column_name(df):
    old_columns = df.columns
    new_columns = [c.replace(' ', '_') for c in old_columns]
    return df.toDF(*new_columns)
```

```python
for d in df.collect():
    temp = spark.read.format("csv").option("header","true").load(f"s3://proj-nova/Retail_sales/{d.filename}.{d.extension}")
    temp = normalize_column_name(temp)
    temp.write.mode("overwrite").saveAsTable(f'01_bronze.default.{d.table_name}')
    print(f"Table {d.table_name} created")
```
