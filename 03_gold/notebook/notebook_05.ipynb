**Creation of data cube with total sales and profit**

```python
from pyspark.sql.functions import sum
 
df = spark.read.table("03_gold.default.fact_sales")
 
df_cube = df.cube(
    "date",
    "store_id",
    "product_id"
).agg(
    sum("total_sales").alias("total_sales"),
    sum("total_profit").alias("total_profit")
)
 
spark.sql("DROP TABLE IF EXISTS 03_gold.default.sales_cube")
 
df_cube.write.format("delta").mode("overwrite").saveAsTable("03_gold.default.sales_cube")
```
