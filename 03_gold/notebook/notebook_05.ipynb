```python
spark.sql("USE CATALOG 03_gold")
spark.sql("USE SCHEMA default")
from pyspark.sql.functions import (
    col, initcap, regexp_replace, to_date, coalesce, 
    current_timestamp, expr, trim,year,month,dayofmonth,date_format,sum
)
```

**Creation of data cube with total sales and profit**

```python
df = spark.read.table("03_gold.default.fact_sales")
df_cube = df.cube(
    "date",
    "store_id",
    "product_id"
).agg(
    sum("total_sales").alias("total_sales"),
    sum("total_profit").alias("total_profit")
)
df_cube.write.mode("overwrite").saveAsTable("03_gold.default.sales_cube")
```
